### _imports_

```python
import datetime
import itertools
import pandas as pd
import numpy as np

from pprint import pprint
from collections import Counter
```


##  1. Археологические находки

### 1.1

```python
# 1 Загрузка, общее число записей

arch_findings_xls = 'data-54581-2018-08-28.xlsx'

arch_findings_df = pd.read_excel(arch_findings_xls)

print( 'Общее количество находок: {}'.format(len(arch_findings_df)) )
```

###  1.2

```python
# 2 Добавляем колонки Год и Месяц

discovery_date = arch_findings_df['DateOfDiscovery'].apply(lambda dt: datetime.datetime.strptime(dt, '%d.%m.%Y'))

arch_findings_df = arch_findings_df.join(
    pd.DataFrame.from_records(list(discovery_date.apply(lambda dt: (dt.year, dt.month)).values),
    columns = ['YearOfDiscovery', 'MonthOfDiscovery']))

arch_findings_df.head()
```

### 1.3

```python
# 3 Распределение по годам и по годам / месяцам

print('Количество находок по годам:')
print( arch_findings_df['YearOfDiscovery'].value_counts().sort_index() )
print()
print('Количество находок по годам / месяцам:')
print( arch_findings_df.groupby('YearOfDiscovery')['MonthOfDiscovery'].value_counts().sort_index() )
```

### 1.4

```python
# 4 Запись с самым длинным описанием

description = arch_findings_df['DescriptionOfFinding'].str;

max_descr_len = arch_findings_df[description.len() == description.len().max()]

print( 'Запись с самым длинным описанием ({}):'.format(max_descr_len['DescriptionOfFinding'].str.len().values[0]) )
print(max_descr_len.values[0][1])
max_descr_len
```

### 1.5 (bonus)

```python
# 5 Счётчик количества слов в описаниях находок
#
# Можно было бы сделать более сложную очистку (по примеру заданий из пары предшествующих домашних работ).
# Но даже на таком уровне видно, что наибольшее количество находок относится к 17 и 18 векам, а по локации - Зарядье.
#
# *Правда, это не совсем то, что можно было бы назвать "с использованием pandas", но тем не менее)
#

Counter(list(itertools.chain(*[ text.lower().split() for text in arch_findings_df['DescriptionOfFinding'] ]))).most_common()[:20]
```

###  2.1

```python
# 1 Загрузка, общее число записей

sports_grounds_xls = 'data-25342-2018-08-17.xlsx'

sports_grounds_df = pd.read_excel(sports_grounds_xls)

print( 'Количество спортплощадок: {}'.format(len(sports_grounds_df)) )
```

###  2.2

```python
# 2 Выводим все колонки

pd.set_option('display.max_columns', None) 

sports_grounds_df.head()
```

### 2.3

```python
# 3 Удаляем строчки с NaN

print('Число строк после DataFrame.dropna(): {}'.format(len(sports_grounds_df.dropna(axis = 0))))

# Причина в том, что в любой строке есть какое-либо пустое поле, например:
sports_grounds_df[~(sports_grounds_df['TechServiceComments'].isnull() | sports_grounds_df['HelpPhoneExtension'].isnull())]
```

###  2.4

```python
# 4 Бесплатные площадки с удобствами для инвалидов

disability_friendly_df = sports_grounds_df[(sports_grounds_df['Paid'] == 'бесплатно') & 
                                           ~((sports_grounds_df['DisabilityFriendly'] == 'не приспособлен') |
                                             (sports_grounds_df['DisabilityFriendly'].isnull()))]

print('Для инвалидов приспособлены {} площадок из {} ({:5.2f}%)'.format(len(disability_friendly_df), len(sports_grounds_df), 100 * len(disability_friendly_df)/len(sports_grounds_df)))
print()
print( sports_grounds_df['DisabilityFriendly'].value_counts(dropna = False) )

disability_friendly_df.head()
```
