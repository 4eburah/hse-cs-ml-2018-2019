```python
import pandas as pd
df = pd.read_excel('data-54581-2018-08-28.xlsx')
len(df)
```

```python
df['MonthOfDiscovery'] = df['DateOfDiscovery'].apply(lambda x: x.split('.')[1])
df['YearOfDiscovery'] = df['DateOfDiscovery'].apply(lambda x: x.split('.')[2])
```

```python
df['YearOfDiscovery'].value_counts()
```

```python
df['MonthOfDiscovery'].value_counts()
```

```python
# Нахождение самого длинного описания в 1 строчку:
df['DescriptionOfFinding'].iloc[df['DescriptionOfFinding'].str.len().idxmax()]
```

```python
# Альтернативный вариант:
df['DescriptionOfFindingLength'] = df['DescriptionOfFinding'].apply(len)
max_len = df['DescriptionOfFindingLength'].max()
df[df['DescriptionOfFindingLength'] == max_len]['DescriptionOfFinding'].values[0]
```
