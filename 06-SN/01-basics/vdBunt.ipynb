```python
import pandas as pd
import numpy as np
import networkx as nx

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
adjacency_matrix = pd.read_csv('vdBunt_data/VRND32T5.DAT', sep=' ', header=None, skipinitialspace=True)

pd.set_option('display.max_columns', len(adjacency_matrix.columns))

adjacency_matrix
```

```python
# удаление строк, состоящих только из 9-ток - actor non-response
# adjacency_matrix = vrnd32t5.mask(9 == digraph_data).dropna(0, 'all')
```

```python
G = nx.DiGraph()

G.add_nodes_from(adjacency_matrix.columns)

for row in adjacency_matrix.index:
    for col in adjacency_matrix.columns:
        rel = adjacency_matrix.loc[row, col]
        if rel not in [0, 6, 9]: #  0 = unknown, 6 = item non-response, 9 = actor non-response
            G.add_edge(row, col, length = rel)
```

```python
personal_data = pd.read_csv('vdBunt_data/VARS.DAT', sep=' ', header=None, names=['gender','program','smoking'], skipinitialspace=True)

personal_data
```

```python
in_dc_total = np.array(list(nx.in_degree_centrality(G).values()))
out_dc_total = np.array(list(nx.out_degree_centrality(G).values()))
```

```python
def report_degree_centrality(in_dc, out_dc, group):
    print('Mean IN degree centrality ({}): {:.3f}'.format(group, np.mean(in_dc)))
    print('Mean OUT degree centrality ({}): {:.3f}'.format(group, np.mean(out_dc)))
    plt.hist([in_dc, out_dc])
    plt.xlabel('degree centrality')
    plt.title('IN and OUT degree centralities: {}'.format(group))
    plt.legend(['IN', 'OUT']);
```

```python
report_degree_centrality(in_dc_total, out_dc_total, 'total')
```

```python
girls_ix = personal_data[1 == personal_data.gender].index
boys_ix = personal_data[2 == personal_data.gender].index
```

```python
in_dc_girls = in_dc_total[girls_ix]
out_dc_girls = out_dc_total[girls_ix]

report_degree_centrality(in_dc_girls, out_dc_girls, 'girls')
```

```python
in_dc_boys = in_dc_total[boys_ix]
out_dc_boys = out_dc_total[boys_ix]

report_degree_centrality(in_dc_boys, out_dc_boys, 'boys')
```
